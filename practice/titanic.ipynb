```python
columns_list = list(df.columns)
columns_list
```

```python
missing_value_df = df.isnull()
missing_value_df.head(20)
```

```python
for columns in columns_list:
    print(columns)
    print(missing_value_df[columns].value_counts())
    print(" ")
```

```python
## get the columns with missing values
missing_columns = [col for col in columns_list if missing_value_df[col].any()]
missing_columns
```

```python
for columns in missing_columns:
    print(columns)
    print(missing_value_df[columns].value_counts())
    print(" ")
```
